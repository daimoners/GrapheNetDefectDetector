```python
files = sorted(file_list)
print(files)
```

```python
plt.style.available
```

```python
import random
plt.style.use( 'seaborn-v0_8-paper')
n = 6
random_files = random.sample(files, n)
cols = 2
rows = n // cols
# Prepara il layout della griglia per il plot (2x2 per 4 file)
fig, axs = plt.subplots(rows, cols, figsize=(10, 10))
axs = axs.ravel()  # Rende l'array axs flat per poterci iterare facilmente

# Cicla sui file casuali e crea un plot per ciascuno
for idx, path in enumerate(random_files):
    with open(path) as f:
        file = json.load(f)
        
        ldos = file["LDOS"]  # Dovrebbe essere un dizionario con chiavi come "region1", "region2", ecc.
        
        # Inizializzo un array per contenere le LDOS di tutte le regioni
        ldos_array = []
        energy = ldos["region1"]["energy"]  # Supponiamo che l'energia sia uguale in tutte le regioni
        
        # Ciclo su tutte le regioni e colleziono le LDOS
        for key, value in ldos.items():
            states = value["states"]
            ldos_array.append(states)
        
        # Converti ldos_array in un numpy array
        ldos_array = np.array(ldos_array)
        
        # Calcola la media su tutte le regioni
        ldos_mean = np.mean(ldos_array, axis=0)
        
        # Plotta energy vs ldos_mean in uno degli assi
        axs[idx].plot(energy, ldos_mean,"--",linewidth=1)
        axs[idx].set_title(f"Plot {idx + 1}: {path.stem}")
        axs[idx].set_xlabel("Energy (eV)")
        axs[idx].set_ylabel("LDOS")
        
# Ottimizza lo spacing tra i subplots
plt.tight_layout()
plt.show()
```
